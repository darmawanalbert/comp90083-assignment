```python
# Library Import
from mesa.batchrunner import BatchRunner
from road_network_model.model import RoadNetworkModel

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Mesa BatchRunner

fixed_params = {"number_of_cars": 30,
                "width": 100,
                "height": 100,
                "is_odd_even_policy_enabled": True
               }

variable_params = {"policy_1_start_time": range(6,9),
                   "policy_1_duration": range(1,3),
                   "policy_2_start_time": range(13, 21),
                   "policy_2_duration": range(1,3)
                  }


batch_run = BatchRunner(RoadNetworkModel,
                        variable_parameters=variable_params,
                        fixed_parameters=fixed_params,
                        iterations=5,
                        max_steps=100
                       )

batch_run.run_all()

batch_run_results = batch_run.get_model_vars_dataframe()
batch_run_results.head()
```
